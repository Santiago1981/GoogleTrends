```python
# You must connect to Google. Also, I will import the pandas library to store and visualize the data.
import pandas as pd                        
from pytrends.request import TrendReq
pytrend = TrendReq()
```

```python
pytrend.build_payload(kw_list=["roland garros"])
# Intereses por País según Google Trends
df = pytrend.interest_by_region()
df.head(30)
```

```python
pytrend.build_payload(kw_list=["roland garros 2021"])
# Intereses por País según Google Trends
df = pytrend.interest_by_region()
df.head(30)
```

```python
# The values are calculated on a scale from 0 to 100, where 100 is the location with the most popularity as a fraction of total searches in that location, a value of 50 indicates a location which is half as popular. 
# A value of 0 indicates a location where there was not enough data for this term.
```

```python
df.reset_index().plot(x='geoName', y='maitland ward', figsize=(120, 10), kind ='bar')
```

```python
# Get Google Keyword Suggestions
keywords = pytrend.suggestions(keyword="maitland")
df = pd.DataFrame(keywords)
df.drop(columns= 'mid')   # This column makes no sense
```

```python
# Related Queries
pytrend.build_payload(kw_list=["roland garros 2021"])
# Related Queries, returns a dictionary of dataframes
related_queries = pytrend.related_queries()
related_queries.values()
```

```python
# Related Topics, returns a dictionary of dataframes
pytrend.build_payload(kw_list=["alberto fernandez medidas"])
related_topic = pytrend.related_topics()
related_topic.values()
```
